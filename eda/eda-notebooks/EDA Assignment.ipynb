# Removing Outliners

we will analyse and find the outliers in all the numerical columns with 1.5 IQR rule and we will remove the outlier records respectively


```python
columns_outlier = ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','AGE']
for col in columns_outlier: 
    q1 = inp0[col].quantile(0.25)
    q3 = inp0[col].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    inp0 = inp0.loc[(inp0[col] > range_low) & (inp0[col] < range_high)]

inp0.shape
```

let's find out the percentage of rows removed

```python
round((100 - 100*(len(inp0.index)/307511)),2)
```

we have removed almost 10% of outliners from the dataframe, and now the dataframe is done with the cleaning and removing unwanted rows and columns , so that the analysis of data will be performed efficiently.

```python
# let us define a function to categorizing the AGE column in different categories ie. Young age, Mid age , Old age

def age_group(x):
    if (x < 40):
        return 'Young Age'
    elif (x >= 40 and x < 60):
        return 'Mid Age'
    else:
        return 'Old Age'
```

```python
#now we will create a new column of these catogories of AGE

inp0['AGE_GROUP'] = inp0['AGE'].apply(age_group)
inp0.head()
```

```python
# let us categorize the same thing with AMT_CREDIT

round(inp0.AMT_CREDIT.describe(),2)
```

```python
# Define function for categorizing CREDIT_GROUP (High, Medium and Low)
def credit_group(x):
    if (x < 500000):
        return 'Low'
    elif (x >= 500000 and x < 750000):
        return 'Medium'
    else:
        return 'High'
```

```python
#Creating a new column AGE_GROUP
inp0['CREDIT_GROUP'] = inp0['AMT_CREDIT'].apply(credit_group)
inp0.head()
```

```python
# let us categorize the same thing with AMT_INCOME_TOTAL
inp0.AMT_INCOME_TOTAL.describe()
```

```python
# Define function for categorizing AMT_INCOME_TOTAL (Low, Medium and High)
def income_group(x):
    if (x < 100000):
        return 'Low'
    elif(x >= 100000 and x < 150000):
        return 'Medium'
    else:
        return 'High'
```

```python
# we will create new  column INCOME_GROUP with AMT_INCOME_TOTAL (Low, Medium and High)
inp0['INCOME_GROUP'] = inp0['AMT_INCOME_TOTAL'].apply(income_group)
inp0.head()
```

# Removing Outliners

```python
# Finding outliers in all the numerical columns with 1.5 IQR rule and removing the outlier records 
col_list_outlier = ['AMT_ANNUITY','AMT_APPLICATION','AMT_CREDIT','AMT_GOODS_PRICE','DAYS_DECISION','CNT_PAYMENT']
for col in col_list_outlier: 
    q1 = inp1[col].quantile(0.25)
    q3 = inp1[col].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    inp12 = inp1.loc[(inp1[col] > range_low) & (inp1[col] < range_high)]

inp12.shape
```

```python
#percentage of rows lost 
round((100 - 100*(len(inp12.index)/1670214)),2)
```

As a conclusion we lost 25% of data from the total data
